This notebook creates plots of radio images of AGN. 

Radio data from the DRAGNs sample: https://www.jb.man.ac.uk/atlas/frame.html

Additional data from: Skyview

```python
# loading packages

import numpy as np
import matplotlib.pyplot as plt

# standard astronomy packages
from astropy.io import fits
from astropy import units as u
from astropy.coordinates import SkyCoord
from astropy.wcs import WCS
from reproject import reproject_interp

# accessing data from SkyView
from astroquery.skyview import SkyView
from astroquery.vizier import Vizier

# for 3 colour images
from astropy.visualization import make_lupton_rgb
from astropy.utils.data import get_pkg_data_filename
```

```python
hdul = fits.open('./DRAGNs_data/3C219.fits')
d = hdul[0].data
h = hdul[0].header
```

```python
h
```

```python
name = '3C219'
path = SkyView.get_images(position=name, width=0.1*u.deg, height=0.1*u.deg, survey=['WISE 22'], pixels=[900,900])

hdulist_DSS2 = path[0]
d_DSS2 = hdulist_DSS2[0].data
h_DSS2 = hdulist_DSS2[0].header
```

```python
fig = plt.figure(figsize=(5,4))
ax = fig.add_subplot(1,1,1, projection=WCS(h_DSS2))
plt.imshow(d_DSS2)
#ax.contour(d[0,0,:,:], levels=rms_NVSS*np.array([0.5, 1,2,4]), cmap='Blues_r')
ax.set_xlabel('RA', size=18,family='serif')
ax.set_ylabel('Dec', size=18,family='serif')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

plt.show()
```

```python
rms_NVSS = np.sqrt(np.mean(d[0,0,:,:]**2))



fig = plt.figure(figsize=(11,10))
ax = fig.add_subplot(1,1,1, projection=WCS(h), slices=('x','y',0,0))
ax.imshow(d[0,0,:,:], cmap='Oranges', vmax=np.percentile(d[0,0,:,:],99.9))
ax.contour(d[0,0,:,:], levels=rms_NVSS*np.array([0.5, 1,2,4]), cmap='Blues_r')
ax.set_xlabel('RA', size=18,family='serif')
ax.set_ylabel('Dec', size=18,family='serif')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

plt.show()
```
